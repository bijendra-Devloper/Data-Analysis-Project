```python
 def autopct_format(values):
        def my_format(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{:.1f}%\n({v:d})'.format(pct, v=val)
        return my_format

plt.pie(df['Employment Profile'].value_counts(),labels=df['Employment Profile'].unique(), autopct=autopct_format(df['Employment Profile'].value_counts()))
```

```python
## 10. Existing Customer Behavior:
df.groupby(['Existing Customer', 'Credit Score'])['Existing Customer'].count().unstack('Credit Score').plot(kind='bar', stacked=True, figsize=(8, 8))
```

```python
grouped_data = df.groupby(['Credit Score', 'Existing Customer']).size().unstack(fill_value=0)
```

```python
plt.bar(grouped_data.index, grouped_data['Yes'], label='Yes')

plt.bar(grouped_data.index, grouped_data['No'], bottom=grouped_data['Yes'], label='No')

plt.xlabel('Credit Score')

plt.ylabel('Count')

plt.title('Stacked Bar Chart for Existing Customers by Credit Score')

plt.legend()
```
